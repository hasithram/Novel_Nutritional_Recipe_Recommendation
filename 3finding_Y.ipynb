# 3.FORMULATE Y

```python
import pandas as pd
import ast
from time import time
import numpy as np
```

```python
recipe = pd.read_csv('no_drop.csv')
recipe = recipe.drop(columns=['Unnamed: 0','recipe_name','image_url','cooking_directions'])
recipe.head()
```

```python
#Y gives the nutritional content of the whole recipe present
list_of_dict = []

for row in recipe.nutritions:
    list_of_dict.append(ast.literal_eval(row))
```

```python

fat_list = []

protein_list = []

sodium_list = []

sugars_list = []


for x in range(len(list_of_dict)):
   
    fat_list.append(list_of_dict[x]['fat']['amount'])
    
    protein_list.append(list_of_dict[x]['protein']['amount'])
  
    sodium_list.append(list_of_dict[x]['sodium']['amount'])
   
    sugars_list.append(list_of_dict[x]['sugars']['amount'])

    
```

```python
data = {'sugars': sugars_list, 'fat': fat_list,'protein': protein_list,'sodium': sodium_list}
df = pd.DataFrame(data)
df.index = recipe['recipe_id']
df.head()
```

```python
y=df.values.tolist()
print(y)
```

```python
y=np.array(y)
np.save('y.npy',y)
```
