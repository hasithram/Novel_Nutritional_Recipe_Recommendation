# 1.PREPROCESSING THE RECIPE DATA

```python
#Import the recipe-dataset
import os
import pandas as pd
df = pd.read_csv('core-data_recipe.csv')
df.head()
```

```python
#Step 1 of preprocessing: Remove Bracketed comments from ingredients
import re
filename='core-data_recipe.csv'
re.sub(r'\([^)]*\)', '', filename)
```

```python
#Step 2 of preprocessing: Remove Punctuations from ingredients

df=pd.read_csv('core-data_recipe.csv')
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","_",
              "`","{","|","}","~","–","®"]
for char in spec_chars:
    df['ingredients'] = df['ingredients'].str.replace(char, '')
```

```python
df.head()
```

```python
#Replace ^ with , as a seperator between ingredients
spec=["^"]

for char in spec:
    df['ingredients'] = df['ingredients'].str.replace(char,',')
```

```python
df.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
#Step 3 of preprocessing: Retain only Nouns in ingredients
from nltk import pos_tag
for i in range(0,len(df['ingredients'])):
    tagged_ing_titles = df['ingredients'].iloc[i].split(",")
    s=""
    for x in tagged_ing_titles:
        k=pos_tag(x.split())
        for y in k:
            if "NN" in y[1]:
                s+=y[0]
                s+=" "
        s=s[:-1]+","
    s=s[:-1]
    df['ingredients'].iloc[i]=s.lower()
```
